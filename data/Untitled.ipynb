```python
class NearEarthObject(object):
    """
    Object containing data describing a Near Earth Object and it's orbits.

    # TODO: You may be adding instance methods to NearEarthObject to help you implement search and output data.
    """

    def __init__(self, **kwargs):
        """
        :param kwargs:    dict of attributes about a given Near Earth Object, only a subset of attributes used
        """
        # TODO: What instance variables will be useful for storing on the Near Earth Object?
        # id, name, nasa_jpl_url, absolute_magnitude_h, estimated_diameter_min_kilometers
        print(kwargs.items())
        for key, value in kwargs.items():
            if key == 'id':
                self.id = value
            elif key == 'name':
                self.name = value
            elif key == 'nasa_jpl_url':
                self.nasa_jpl_url = value
            elif key == 'absolute_magnitude_h':
                self.absolute_magnitude_h = value
            elif key == 'estimated_diameter_min_kilometers':
                self.estimated_diameter_min_kilometers = value
        self.list_of_orbits = []

    def update_orbits(self, orbit):
        """
        Adds an orbit path information to a Near Earth Object list of orbits

        :param orbit: OrbitPath
        :return: None
        """
        # TODO: How do we connect orbits back to the Near Earth Object?
        self.list_of_orbits.append(orbit)

    def __str__(self):
        message = 'Name: ' + str(self.name) \
                 + ', NASA_Jpl_Url: ' + str(self.nasa_jpl_url) \
                 + ', Absolute_Magnitude_H: ' + str(self.absolute_magnitude_h) \
                 + ', Estimated_Diameter_Min_Kilometers: ' + str(self.estimated_diameter_min_kilometers)
        return message

    
        


class OrbitPath(object):
    """
    Object containing data describing a Near Earth Object orbit.

    # TODO: You may be adding instance methods to OrbitPath to help you implement search and output data.
    """

    def __init__(self, **kwargs):
        """
        :param kwargs:    dict of attributes about a given orbit, only a subset of attributes used
        """
        # TODO: What instance variables will be useful for storing on the Near Earth Object?
        #name, miss distance in km, and orbit date
        for key, value in kwargs.items():
            if key == 'name':
                self.name = value
            elif key == 'miss_distance_kilometers':
                self.miss_distance_kilometers = value
            elif key == 'close_approach_date':
                self.close_approach_date = value

    def __str__(self):
        message = 'Name: ' + str(self.name) \
                 + ', Miss_Distance_Kilometers: ' + str(self.miss_distance_kilometers) \
                 + ', Close_Approach_Date: ' + str(self.close_approach_date)
        return message
```

```python


def generate_neo_object(val):
    """
    Generates Neo object from every record of the csv file
    """
    id_, name, nasa_jpl_url, absolute_magnitude_h, estimated_diameter_min_kilometers, close_approach_date, miss_distance_kilometers = val
    neo = NearEarthObject(id = id_, \
                          name = name, \
                          nasa_jpl_url = nasa_jpl_url, \
                          absolute_magnitude_h = absolute_magnitude_h, \
                          estimated_diameter_min_kilometers = estimated_diameter_min_kilometers
                         )
    return neo


def generate_orbit_path(val):
    """
    Generates Neo object from every record of the csv file
    """
    id, name, nasa_jpl_url, absolute_magnitude_h, estimated_diameter_min_kilometers, close_approach_date, miss_distance_kilometers = val
    return OrbitPath(name = name, \
                     miss_distance_kilometers = miss_distance_kilometers, \
                     close_approach_date = close_approach_date)


df_sub = df[['id', 'name', 'nasa_jpl_url',\
                               'absolute_magnitude_h', 'estimated_diameter_min_kilometers',\
                               'close_approach_date', 'miss_distance_kilometers']]
df_sub['neo_object'] = df_sub.apply( generate_neo_object , axis = 1)
```

```python
df_sub['orbit_part_object'] = df[['id', 'name', 'nasa_jpl_url',\
                               'absolute_magnitude_h', 'estimated_diameter_min_kilometers',\
                               'close_approach_date', 'miss_distance_kilometers']].apply( generate_orbit_path , axis = 1)

df_res = pd.DataFrame([])
df_res['orbit_part_v1'] = df_sub.groupby('name')['orbit_part_object'].apply(list)
df_res.reset_index(inplace=True)

x = pd.DataFrame([])
x['neo_object_v1'] = df_sub.groupby('name')['neo_object'].first()
x = x.reset_index()

df_res = pd.merge(df_res, x, on='name')
df_res.head()
```

```python
def final_neo_object( val):
    neo_object_v1, orbit_part_v1 = val
    #print(neo_object_v1, orbit_part_v1)
    for each in orbit_part_v1:
        neo_object_v1.update_orbits(each)

    return neo_object_v1

df_res['neo_object_v2'] = df_res[['neo_object_v1', 'orbit_part_v1']].apply(final_neo_object, axis = 1)

df_res.head()
```
